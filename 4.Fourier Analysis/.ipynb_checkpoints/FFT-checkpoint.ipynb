## Fast Fourier Transformation

Since computational complexity for DFT is $O(N^2)$, it will be inefficient sometimes. And the fast Fourier transformation (FFT) has computational complexity $O(NlogN)$. FFT is the preferred algorithm typically.


```julia
using FFTW
using Plots

# define function to be analysed
function func(t::Number)
    return sin(π*t)+cos(3*π*t)+2*sin(7*π*t)
end

tList = 0.0:0.001:1.999
realSpace = func.(tList)

kSpace = fft(realSpace)
ω = [(i-1)*2*π/2 for i in 1:2000]
kSpaceModule = abs.(kSpace)

plot(ω[1:15], kSpaceModule[1:15], leg=false)
```
