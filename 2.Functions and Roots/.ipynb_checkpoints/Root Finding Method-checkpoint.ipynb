## Bisecion method

When we talk about finding a root, the straightforward way is to find an interval bracketing the root and then divide the interval in half until we find an interval small enough. That's surely a good solution. And it's called bisection method.

### Example:

Let's consider a transcendental equation: $x = \alpha tan(x)$. If you are familiar with the statistical mechanics, you will remember this is the self-consistent equation of Ising model under mean field approximation. Whatever, let's plot the function $f(x)=0.5tan(x)-x$, and then use the bisection method.

## Code Example (Julia language):

```julia
using Plots

x = [i for i in 0:0.001:π/2-0.1]
y = 0.5.*tan.(x)-x
plot(x, y, leg=false)
```

And obviously $x=0$ is not the root we want. We have already known it. The other root is between 1.0 and 1.4.

```julia
function func(x::Float64)
    return 0.5*tan(x)-x
end

function Bisection(f::Function, x₁::Float64, x₂::Float64)
    # the sign of f(x₁), true denotes +, false denotes -
    f_x_sign = false
    if f(x₁) > 0 
        f_x_sign = true
    end
    
    mid = 0
    for i in 1:1000000
        mid = (x₁+x₂)/2
        mid_value = f(mid)
        if mid_value<ϵ && f(mid)>ϵ
            break
        elseif mid_value > 0
            if f_x_sign
                x₁ = mid
            else
                x₂ = mid
            end
        else
            if f_x_sign
                x₂ = mid
            else
                x₁ = mid
            end
        end
    end
    return mid
end

ϵ = 0.0001
root = Bisection(func, 1.0, 1.4)
print("x:\t")
print(root)
print(";\tfunction value:\t")
print(func(root))
```

That's a pretty accurate result.

However, you may notice that the bisection method has many drawbacks.

1. You must find the interval first.
2. The boundary points must have opposite sign.
3. You can only get one root.

Unfortunately, we can't overcome the third drawback.

To overcome the first and the second drawback, we can use Newton-Raphson method.

## Newton-Raphson Method

Using the Taylor's expansion, we have $f(x)\approx f(x_0)+(x-x_0)f\prime(x_0)$, which means $x=x_0-\frac{f(x_0)}{f\prime(x_0)}$

### Example:

We will reconsider the function above.

### Code Example (Julia language):

```julia
function func_derivative(x::Float64)
    return sec(x)^2-1
end

function Newton_Raphson(f::Function, deri::Function, x₀::Float64)
    for i in 1:10000
        function_value = f(x₀)
        derivative_value = deri(x₀)
        
        if abs(function_value) < ϵ
            break
        end
        
        x₀ -= function_value/derivative_value
    end
    
    return x₀
end

ϵ = 0.0001
root = Newton_Raphson(func, func_derivative, 1.4)
print("x:\t")
print(root)
print(";\tfunction value:\t")
print(func(root))
```

Though Newton-Raphson method is faster than bisection method, sometimes Newton-Rephson method is unstable. Now the most robust mothod is Van Wijngaarden-Dekker-Brent method. And we can take a look of it.

## Van Wijngaarden-Dekker-Brent Method

Van Wijngaarden-Dekker-Brent method combines bracketing method with open method, which means it's faster than bracketing method and more reliable than open method.
