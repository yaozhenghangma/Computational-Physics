## Associated Legendre Polynomials

The associated Legendre functions satisfy the recurrence relation:
$$(l-m)P^m_l=x(2l-1)P^m_{l-1}-(l+m-1)P^m_{l-2}$$
which is easier to be programmed than other expressions.

Since the diagonal terms satisfy:
$$P^l_l=(-1)^l*(2l-1)!!(1-x^2)^{l/2}$$
where $n!!$ denotes the product of all odd integers less or equal to n,
we can get $P^m_m$ and then $P^m_{m+1}$ ... and finally $P^m_l$.

Besides, the functions of negative m are proportional to those of positive m:
$$P^{-m}_l=(-1)^m\frac{(l-m)!}{(l+m)!}P^m_l$$

#### Example(Julia Language):

```julia
using Plots

function LegendreP(l, m, x)
    if l < 0
        return 0
    end
    
    if m < 0
        m = abs(m)
        return (-1)^m*factorial(l-m)/factorial(l+m)*LegendreP(l, m, x)
    else
        pmm = (-1)^m*oddProduct(2*m-1)*(1-x^2)^(m/2)
        if m == l
            return pmm
        elseif (m+1) == l
            pmm1 = x*(2*m+1)*pmm
            return pmm1
        else
            pmm1 = x*(2*m+1)*pmm
            for i in (m+2):1:l
                pmm2 = x*(2*i-1)*pmm1/(i-m) - (i+m-1)*pmm/(i-m)
                pmm = pmm1
                pmm1 = pmm2
            end
            return pmm2
        end
    end
end
    
function oddProduct(n)
    if mod(n, 2) == 0
        return productRecurrence(n-1)
    else
        return productRecurrence(n)
    end
end

function productRecurrence(n)
    if n>1
        return n*productRecurrence(n-2)
    else
        return 1
    end
end

x = -1.0:0.02:1.0
y = LegendreP.(2, 1, x)
plot(x, y, leg=false)
```
