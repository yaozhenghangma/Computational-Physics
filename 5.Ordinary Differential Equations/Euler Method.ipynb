## Euler's Method:

One obvious way to solve differential equation is using Taylor expansion truncated to two terms.

That is Euler's method: $y_{i+1} = y_{i} + y\prime*h$

### Example:
Solve $y\prime = y^2+1$, with 0<x<1, y(0)=0.

For comparision, h=0.05, 0.10, 0.15 and 0.20.

#### Code Example (Julia language):

```julia
using Plots

# function to get derivative
function func(y::Float64)
  return y^2+1
end

# function to solve differential function f with interval h and initial condition y0
# return graphing
function Euler(f::Function, h::Float64, y0::Float64)
    x = [i for i in 0:h:1.0]
    y = [0.0 for i in 0:h:1.0]
    y[1] = y0

    for i in 1:(size(y)[1]-1)
        y[i+1] = f(y[i])*(x[i+1]-x[i])+y[i]
    end

    plot!(x, y, label="h="*string(h))
end

# plot the analytical result
plot(x -> tan(x), 0:0.01:1.0, label="real")

# plot approximated result
Euler(func, 0.05, 0.0)
Euler(func, 0.1, 0.0)
Euler(func, 0.15, 0.0)
Euler(func, 0.2, 0.0)
```

For Euler method, the local error is obviously $O(h^2)$, which means global error is $O(h)$.

## Midpoint Mothed(modified Euler's method):

While the Euler's method use the derivative at beginning of the interval 
as the derivative over the entire step, the end of the interval is not used 
which usually leads to a bad approximation.

To solve this problem, we use the derivative of midpoint of the interval. 
That is the modpoint method.

$$y(x_{mid}) = y_0 + \frac{h}{2}y_0\prime$$
$$y(x_0+h) = y(x_0)+hf(x_{mid}, y_{mid})$$

### Improved Euler's method:

There is another way to solve drawback of Euler's method. We can simply use 
the mean value of the beginning point and the end point, i.e.

$$y(x_0+h) = y(x_0) + h\frac{f(x_0, y_0)+f(x_0+h, y_0+hf_0)}{2}$$

### Example:

Reconsider the problem above. This time we use midpoint method and improved Euler's method.

### Code Example (Julia language):

```julia
# midpoint method
function Midpoint(f::Function, h::Float64, y0::Float64)
    x = [i for i in 0:h:1.0]
    y = [0.0 for i in 0:h:1.0]
    y[1] = y0
    
    for i in 1:(size(y)[1]-1)
        y_mid = y[i] + f(y[i])*h/2
        y[i+1] = y[i] + f(y_mid)*h
    end
    
    plot!(x, y, label="modpoint")
end

# improved Euler's method
function improve_Euler(f::Function, h::Float64, y0::Float64)
    x = [i for i in 0:h:1.0]
    y = [0.0 for i in 0:h:1.0]
    y[1] = y0
    
    for i in 1:(size(y)[1]-1)
        dy_begin = f(y[i])*h
        dy_end = f(y[i]+dy_begin)*h
        y[i+1] = y[i] + (dy_begin+dy_end)/2
    end
    
    plot!(x, y, label="improved Euler")
end

# plot the analytical result
plot(x -> tan(x), 0:0.01:1.0, label="real")

# plot the result of Euler's method
Euler(func, 0.2, 0.0)

# plot the result of midpoint method
Midpoint(func, 0.2, 0.0)

# plot the result of improved Euler's method
improve_Euler(func, 0.2, 0.0)
```

For midpoint method and improved Euler's method, the local error term is $O(h^3)$, which means global error term is $O(h^2)$. You can get this result from Taylor expansion. 

At this time, you may wonder why we don't use $h^2$ in Euler's method to get the same accuracy. The reason is simple. Using midpoint or improved Euler's method, we
coult calculate less function value and less do less multiplication under the same
accuracy.

## Higher Order Differential Equation

By now, we have only discussed 1st order differential equation. However, it's easy to higher order. What we need to do is just vectorization for differential equation.
Let's learn about it through an example.

### Example:

Here, we consider a simple pendulum with string lenght $l=1m$ and initial angle $\theta=\pi/3$. And typically we consider this pendulum on earth. So the standard 
gravity is $9.8m/s^2$

So how do we do vectorization? Wow, that's simple. We can define a vector:

$$\vec{y}=\big(
\begin{matrix} 
\theta \\
\frac{d\theta}{dt} 
\end{matrix}\big)
$$

and:

$$\vec{f} = \big(
\begin{matrix}
\frac{d\theta}{dt} \\
-\frac{g}{l}sin\theta
\end{matrix}\big)
$$

Thus the 2ed order differential equation becomes $\vec{f}=\vec{y}\prime$, which is 1st order differential equation. And we could solve it using modpoint method.

### Code Example (Julia language):

```julia
function f(y::Array{Float64})
    vector_f = similar(y)
    vector_f[1] = y[2]
    vector_f[2] = -g/l*sin(y[1])
    return vector_f
end

function Midpoint(f::Function, h::Float64, θ₀::Float64)
    t = [i for i in 0:h:5*π]
    y = Array{Float64}(undef, 2, size(t)[1])
    y[1, 1] = θ₀
    y[1, 2] = 0
    
    for i in 1:(size(y)[2]-1)
        y_mid = y[:, i] + f(y[:, i]).*h./2
        y[:, i+1] = y[:, i] + f(y_mid).*h
    end
    
    plot(t, y[1, :], label="modpoint")
end

g = 9.8
l = 1
Midpoint(f, 0.1, π/3)
```

Thus we get the approximation for simple pendulum movement.

However, you may notice that the energy seems not conservation value, for the angle is larger and larger. That's impossible! How can we get a rediculous result like that? Don't worry. That's just a drawback of our algorithm. This drawback will show up in any periodic problem. And we can modify Euler's method to get the corrent result.
