## Interpolation

Since Lagrange interpolation and Newton interpolation use high degree polynomial, in some place, there will be large interpolation value. To solve this problem, we consider to interpolation in every two points.

### Spline Interpolation

Typically, there are three kinds of spline interpolation, linear, quadratic and cubic.

1. Linear Spline:$f_i(x)=a_ix+b_i$
    * $f_i(x_i) = y_i$
    * $f_i(x_{i+1}) = y_{i+1}$
2. Quadratic Spline:$f_i(x)=a_ix^2+b_ix+c_i$
    * $f_i(x_) = y_i$
    * $f_i(x_{i+1}) = y_{i+1}$
    * $f_i^\prime(x_{i+1})=f_{i+1}^\prime(x_{i+1})$
    * $f_1^{\prime\prime}=0$
3. Cubic Spline:$f_i(x)=a_ix^3+b_ix^2+c_ix+d_i$
    * $f_i(x_) = y_i$
    * $f_i(x_{i+1}) = y_{i+1}$
    * $f_i^\prime(x_{i+1})=f_{i+1}^\prime(x_{i+1})$
    * $f_i^{\prime\prime}(x_{i+1})=f_{i+1}^{\prime\prime}(x_{i+1})$
    * $f_1^{\prime\prime}=0$
    * $f_{n-1}^{\prime\prime}=0$
    
#### Example:

Let's reconsider function $f(x)=\frac{1+tanh2\alpha x}{2}$ using three kinds of spline interpolation.

Since the equation can be write as matrix form, we use Gaussian elimination to deal with the matrix.

#### Code Example (Julia language):

```julia
using Plots
include("Elimination.jl")

# define function
function func(x::Float64)
    α = 1.0
    return (1+tanh(2*α*x))/2
end

# get 21 points
x = [i for i in -5.0:0.5:5.0]
y = func.(x)

# Linear Spline
matrixX = Array{Float64}(undef, 40, 40) 
matrixY = Array{Float64}(undef, 40)
matrixX .= 0.0
matrixY .= 0.0
for i in 1:2:40
    matrixX[i, i] = x[Int((i+1)/2)]
    matrixX[i, i+1] = 1.0
    matrixX[i+1, i] = x[Int((i+3)/2)]
    matrixX[i+1, i+1] = 1.0
    matrixY[i] = y[Int((i+1)/2)]
    matrixY[i+1] = y[Int((i+3)/2)]
end
GaussianElimination!(matrixX, matrixY)
coefficient = [matrixY[i]/matrixX[i,i] for i in 1:40]

#plot the figure
xInterpolate = []
yInterpolate = []
for i in 1:20
    xList = x[i]:0.01:x[i+1]
    for x₀ in xList
        push!(xInterpolate, x₀)
        push!(yInterpolate, x₀*coefficient[2*i-1]+coefficient[2*i])
    end
end
plot(xInterpolate, yInterpolate, leg=false)
```

```julia
# Quadratic Spline
matrixX = Array{Float64}(undef, 60, 60) 
matrixY = Array{Float64}(undef, 60)
matrixX .= 0.0
matrixY .= 0.0

# initialize matrixX and matrixY
for i in 1:3:60
    matrixX[i, i] = x[Int((i+2)/3)]^2
    matrixX[i, i+1] = x[Int((i+2)/3)]
    matrixX[i, i+2] = 1.0
    matrixX[i+1, i] = x[Int((i+5)/3)]^2
    matrixX[i+1, i+1] = x[Int((i+5)/3)]
    matrixX[i+1, i+2] = 1.0
    matrixY[i] = y[Int((i+2)/3)]
    matrixY[i+1] = y[Int((i+5)/3)]
    if i == 58
        matrixX[i+2, i] = 2.0
    else
        matrixX[i+2, i] = 2*x[Int((i+5)/3)]
        matrixX[i+2, i+1] = 1.0
        matrixX[i+2, i+3] = -2*x[Int((i+5)/3)]
        matrixX[i+2, i+4] = -1.0
    end
end

# Gaussian Elimination
GaussianElimination!(matrixX, matrixY)
coefficient = [matrixY[i]/matrixX[i,i] for i in 1:60]

#plot the figure
xInterpolate = []
yInterpolate = []
for i in 1:20
    xList = x[i]:0.01:x[i+1]
    for x₀ in xList
        push!(xInterpolate, x₀)
        push!(yInterpolate, x₀^2*coefficient[3*i-2] + x₀*coefficient[3*i-1] + coefficient[3*i])
    end
end
plot(xInterpolate, yInterpolate, leg=false)
```

```julia
using LinearAlgebra
# Cubic Spline
matrixX = Array{Float64}(undef, 80, 80) 
matrixY = Array{Float64}(undef, 80)
matrixX .= 0.0
matrixY .= 0.0

# initialize matrixX and matrixY
for i in 1:4:80
    matrixX[i, i] = x[Int((i+3)/4)]^3
    matrixX[i, i+1] = x[Int((i+3)/4)]^2
    matrixX[i, i+2] = x[Int((i+3)/4)]
    matrixX[i, i+3] = 1.0
    matrixX[i+1, i] = x[Int((i+7)/4)]^3
    matrixX[i+1, i+1] = x[Int((i+7)/4)]^2
    matrixX[i+1, i+2] = x[Int((i+7)/4)]
    matrixX[i+1, i+3] = 1.0
    matrixY[i] = y[Int((i+3)/4)]
    matrixY[i+1] = y[Int((i+7)/4)]
    
    if i == 77
        matrixX[i+2, i] = 6*x[Int((i+7)/4)]
        matrixX[i+2, i+1] = 2.0
    else
        matrixX[i+2, i] = 3*x[Int((i+7)/4)]^2
        matrixX[i+2, i+1] = 2*x[Int((i+7)/4)]
        matrixX[i+2, i+2] = 1.0
        matrixX[i+2, i+4] = -3*x[Int((i+7)/4)]^2
        matrixX[i+2, i+5] = -2*x[Int((i+7)/4)]
        matrixX[i+2, i+6] = -1.0
    end
    
    if i == 1
        matrixX[i+3, i] = 6*x[Int((i+3)/4)]
        matrixX[i+3, i+1] = 2.0
    else
        matrixX[i+3, i] = 6*x[Int((i+3)/4)]
        matrixX[i+3, i+1] = 2.0
        matrixX[i+3, i-4] = -6*x[Int((i+3)/4)]
        matrixX[i+3, i-3] = -2.0
    end
end

# Gaussian Elimination
GaussianElimination!(matrixX, matrixY)
coefficient = [matrixY[i]/matrixX[i,i] for i in 1:80]

#plot the figure
xInterpolate = []
yInterpolate = []
for i in 1:20
    xList = x[i]:0.01:x[i+1]
    for x₀ in xList
        push!(xInterpolate, x₀)
        push!(yInterpolate, x₀^3*coefficient[4*i-3] + x₀^2*coefficient[4*i-2] + x₀*coefficient[4*i-1] + coefficient[4*i])
    end
end
plot(xInterpolate, yInterpolate, leg=false)
```
