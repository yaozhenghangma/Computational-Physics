## Interpolation

Sometimes we want to get more information from some exact points. We need a curve going through these points. That's when interpolation is needed.

### Lagrange Interpolation

Assuming we have n+1 points, we can expand an arbitrary function to a degree n polynomial, $f(x)=\Sigma a_ix^i$. Since $f(x_j)=y_j$, we can rewrite the formula as $f(x)=\sum_{j=0}^{n} A_{j}(x) y_{j} $, where $ A_{j}(x)=\prod_{i=0 \atop i \neq j}^{n} \frac{x-x_{i}}{x_{j}-x_{i}}$. 

#### Example:

Let's consider an interesting function $f(x) = \frac{1+tanh2\alpha x}{2}$. As $\alpha\to\infty$, f(x) becomes zero for $x<0$ and unity for $x>0$. That is, it tends toward the Heaviside step function. Because of that, this function can be used as activation function in machine learning.

#### Code Example (Julia language):

Assuming $\alpha=1.0$, use 21 points to interpolate the function.

```julia
using Plots

# define function
function func(x::Float64)
    α = 1.0
    return (1+tanh(2*α*x))/2
end

# get 21 points
x = [i for i in -5.0:0.5:5.0]
y = func.(x)

# Lagrange interpolation process
function interpolate(x, y, x₀::Float64)
    y₀ = 0.0
    
    for j in 1:size(x)[1]
        A = 1.0
        
        for i in 1:size(x)[1]
            if i == j
                continue
            else
                A *= (x₀-x[i]) / (x[j]-x[i])
            end
        end
        
        y₀ += A * y[j]
    end
    
    return y₀
end

# interpolation
xInterpolate = -4.2:0.05:4.2
yInterpolate = [interpolate(x, y, x₀) for x₀ in xInterpolate]

# plot
plot(xInterpolate, yInterpolate, t=:scatter, markersize=1.0, leg=false)
plot!(-5.0:0.01:5.0, func.(-5.0:0.01:5.0))
```
